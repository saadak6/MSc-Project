```python
import pandas as pd
import datetime
from datetime import datetime, timezone
```

```python
#dam_df = pd.read_csv('turkey_damage.csv', encoding = 'utf-8')
dam_df = pd.read_csv('turkey_damage.csv', encoding='ISO-8859-1')
dam_df

# There was an encoding problem which was faced when we dodn't use the encoding value.The solution was found thorugh:
# https://stackoverflow.com/questions/22216076/unicodedecodeerror-utf8-codec-cant-decode-byte-0xa5-in-position-0-invalid-s
```

```python
dam_df = dam_df.drop(columns = ['Dis No', 'Country', 'Disaster Subtype', 'Region', 'No Affected', 'Seq', 'Glide','Event Name', 'ISO', 'Continent','Origin', 'Associated Dis', 'Associated Dis2', 'OFDA Response','Appeal', 'Declaration', "AID Contribution ('000 US$)", 'River Basin','End Year', 'End Month','End Day', 'CPI','Adm Level', 'Admin1 Code', 'Admin2 Code', 'Geo Locations'])
# DROP columns
```

```python
dam_df
```

```python
na_values = dam_df['No Injured'].isna() & dam_df['No Homeless'].isna()
# check for multiple nan values
```

```python
dam_df[na_values]
```

```python
dam_df = dam_df.dropna(subset = ['No Injured', 'No Homeless'], how = 'all')
#drop the NAN values
```

```python
dam_df
```

```python
dam_df = dam_df.drop(columns = ["Reconstruction Costs ('000 US$)", "Reconstruction Costs, Adjusted ('000 US$)", "Insured Damages ('000 US$)", "Insured Damages, Adjusted ('000 US$)", "Total Damages ('000 US$)"])
# drop more columns
```

```python
#df_new = df_new.sort_values('Year')
x = dam_df['No Injured'].isnull()
dam_df[x]

# check for single NAN values
```

```python
dam_df = dam_df.sort_values('Year')
# sort the values by YEAR
```

```python
x = dam_df['No Homeless'].isnull()
dam_df[x]

# Check NAN values in Homeless
```

```python
nan_fill = dam_df[(dam_df['Year']<1944) & (dam_df['No Injured']<4000)]
nan_fill

# use the same concept as used in Japan damage earthquake cleaning
```

```python
mean_homeless = nan_fill['No Homeless'].mean()
mean_homeless
```

```python
dam_df['No Homeless'] = dam_df['No Homeless'].fillna(mean_homeless)
```

```python
y = dam_df['No Injured'].isnull()
dam_df[y]

# check nam for injured
```

```python
Injured = dam_df[(dam_df['Dis Mag Value']<7)]
# use the same concept for magnitude filtering 

injured_avg = Injured['No Injured'].mean().round(0)
# take the mean and round off its value.
```

```python
injured_avg
# Inhjured average 
```

```python
dam_df['No Injured'] = dam_df['No Injured'].fillna(injured_avg)
# Fill the injured nan values
```

```python
## Do the same thing for total deaths
deaths1 = dam_df[(dam_df['Dis Mag Value']<7)]
deaths2 = dam_df[(dam_df['Dis Mag Value']>=7)]
```

```python
death1_avg = deaths1['Total Deaths'].mean().round(0)
death2_avg = deaths2['Total Deaths'].mean().round(0)
# take mean and round it off
```

```python
deaths1['Total Deaths'] = deaths1['Total Deaths'].fillna(death1_avg)
# fill the nan values
```

```python
deaths2['Total Deaths'] = deaths2['Total Deaths'].fillna(death2_avg)
```

```python
df_merged = pd.concat([deaths1, deaths2], axis = 0)
df_merged
# merge the dataframes
```

```python
df_merged.dropna(subset=['Latitude', 'Longitude'], inplace=True)
```

```python
# now the last time for cost as same as Japan
```

```python
cost1 = df_merged[(df_merged['Dis Mag Value']<7)]
cost2 = df_merged[(df_merged['Dis Mag Value']>=7)]
```

```python
print(cost1['Year'].count())
print(cost2['Year'].count())
```

```python
cost1_avg = cost1["Total Damages, Adjusted ('000 US$)"].mean().round()
print('cost1 average is: ', cost1_avg)
cost2_avg = cost2["Total Damages, Adjusted ('000 US$)"].mean().round()
print('cost2 average is: ', cost2_avg)
```

```python
cost1["Total Damages, Adjusted ('000 US$)"] = cost1["Total Damages, Adjusted ('000 US$)"].fillna(cost1_avg)
cost2["Total Damages, Adjusted ('000 US$)"] = cost2["Total Damages, Adjusted ('000 US$)"].fillna(cost2_avg)
```

```python
final_df = pd.concat([cost1, cost2], axis = 0)
```

```python
final_df
```

```python
final_df = final_df.drop(columns = ['Total Affected'])
# similar to Japan we will drop total affected and then create it again
```

```python
final_df['Total Affected'] = final_df['Total Deaths'] + final_df['No Injured'] + final_df['No Homeless']
```

```python
cols = list(final_df.columns)
# Make our headings into a list

final_df = final_df[cols[0:17] + [cols[-1]] + [cols[17]]]
# Through this we can rearrange our columns to that total affected is not at the end
```

```python
final_df = final_df.sort_values('Year')
# Sort the dataframe by year
```

```python
final_df = final_df.drop(columns = ['Disaster Subsubtype'])

final_df = final_df.drop(columns = ['Local Time'])
#Drop Local time and Disaster Subsubtype as they are not useful to us
```
