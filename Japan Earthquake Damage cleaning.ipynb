#### We still have many values which are NAN. Our goal is to make sure the dataset doesn't have any missing values. This will be tackled in 2 steps:

#### 1). Do some research about the earthquakes and fill the dataset manually. This is a base step in Data Collection. Since the data dates back to 1900, it is common for some important values to be missing. This can be due to data not being recorded, or the data was lost in history.

#### 2). The second step is to fill any missing values which we cannot find through self research. We will use statistical tools to generate average values based on factors such as deaths, magnitude and year.

```python
df_new = pd.read_csv('japan_damage_updated1.csv')
```

#### NOTE: We manually updated some values in japan_damage_updated.csv file. We saved it as japan_damage_updated1.csv because if we overwrote it here then it could cause issues in our code. So to make it simple we continued our work in a seperate file.

```python
df_new.head()
```

```python
df_new = df_new.sort_values('Year')
```

#### Checking for any NAN values and filling them through Pandas.

```python
x = df_new['No Injured'].isnull()
df_new[x]

# we just created a value 'x' to check all the missing values.
```

```python
fill_nan1 = df_new[(df_new['Year']<1944) & (df_new['No Injured']<4000)]
fill_nan1

# This will make a new dataframe which will only show rows which have
# year < 1944 & No injured < 4000
```

### NOTE: The reason we created the filter of year before 1944 and injured less than 4000 is because we see the data and the only NAN vallues for injured belong to the eariest years in our dataset. I did not want to take an average of the whole dataset as it could create issues with the mean value, which is why I took the mean with the condition described so it would make sure the dataset has some normality.  If we were to remove the second condition then the mean will be almost 18000 due to the 1923 earthquake which injured over 103000 people.


```python
mean_injuries = fill_nan1['No Injured'].mean()
mean_injuries
# This gives us our mean injuries which is 1099.2 and we used this value and filled it with the nan values for injured.
```

```python
df_new['No Injured'] = df_new['No Injured'].fillna(mean_injuries)
# The mean injuries have filled nan values of no injured
```

```python
x = df_new['No Injured'].isnull()
df_new[x]
#This shows that we filled all nan values for No of Injured.
```

### Now we will do the same for all remaining NAN values. We will either take average of the whole column or take average with certain conditions so that we can get a more accurate result.

```python
cost_na_check = df_new["Total Damages, Adjusted ('000 US$)"].isna()
# make new df only showing NA values
df_new[cost_na_check]
# show the DF with NA only for total damages, which is the cost after an earthquake.
```

```python
dam_df['Dis Mag Value'].value_counts()
```

```python
cost = df_new[(df_new['Dis Mag Value']<7)]
# make variable cost which is only taking data from magnitude < 7
cost
```

### NOTE: We have to fill the NAN values for cost but we will not take a single average. The reson is cause a lower magnitude (5,6) would have a different cost damage than a magnitude 7 or above. This is why we will create two new variables based on magnitude less than 7, and (7 and above).

```python
cost1 = df_new[(df_new['Dis Mag Value']<7)]
cost2 = df_new[(df_new['Dis Mag Value']>=7)]
# two variables are created based on the condition
```


#### We have created 2 dataframes cost1 and cost2 based on the condition the the magnitude value should be less than 7 and equal or more than 7. Through this we can get better cost average as a higher magnitude might have a higer cost in damages compared to a low magnitude earthquake.

```python
print(cost1['Year'].count())
print(cost2['Year'].count())
# To see if the total is equal to 61 which is True
```

```python
cost1_avg = cost1["Total Damages, Adjusted ('000 US$)"].mean().round()
print('cost1 average is: ', cost1_avg)
cost2_avg = cost2["Total Damages, Adjusted ('000 US$)"].mean().round()
print('cost2 average is: ', cost2_avg)


# We took the average of total damages in both Dataframes cost1 and cost2
```

```python
cost1["Total Damages, Adjusted ('000 US$)"] = cost1["Total Damages, Adjusted ('000 US$)"].fillna(cost1_avg)

# replace the NAN values of cost1 with the average of its table
```

```python
cost2["Total Damages, Adjusted ('000 US$)"] = cost2["Total Damages, Adjusted ('000 US$)"].fillna(cost2_avg)

# replace the NAN values of cost2 with the average of its table
```

#### So now we have filled the Total Damages column by  filtering it through magnitude intensity, we will just concat it and use that dataframe.

```python
#df_merged = cost1.concat(cost2, ignore_index=True)

df_merged = pd.concat([cost1, cost2], axis = 0)

# We will merge the two tables into one. 
# This way the cost values have been updated 
```

```python
df_merged.head() # This is our new dataset which has no NAN values in cost and Injured.
```

```python
test = df_merged["Total Damages, Adjusted ('000 US$)"].isna()
df_merged[test] # We checked and see no NAN values in this dataframe for damages
```

## We will use the same method to fill values for no of Homeless. We will again seperate the table in regards to magnitude

```python
home1 = df_merged[(df_merged['Dis Mag Value']<7)]
home2 = df_merged[(df_merged['Dis Mag Value']>=7)]
# create 2 datadets based on the magnitude filter


home1_avg = home1["No Homeless"].mean().round()
print('Homeless average is: ', home1_avg)
home2_avg = home2["No Homeless"].mean().round()
print('Homeless average is: ', home2_avg)

# We took the average of total damages in both Dataframes cost1 and cost2


home1["No Homeless"] = home1["No Homeless"].fillna(home1_avg)
home2["No Homeless"] = home2["No Homeless"].fillna(home2_avg)

damage_final= pd.concat([home1, home2], axis = 0)
```

### NOTE: The same process was done to fill nan values of homeless using the same concept that low magnitude earthquake would cause lower homelessness compared to a high magnitude earthquake.

```python
test2 = damage_final["No Homeless"].isna()
damage_final[test2]

# all columns have been checked and none have NAN values
```

### NOTE: We need to make one final change in the DataFrame. the column Total Affected is a sum of deaths, injured and homeless. Since we have updated the value of the total Homeless, we will need to recalculate the sum of the three columns. 

#### To make it simple, we will first drop the existing column "Total Affected" and then create  a new one through pandas which will reflect the updated values.

```python
damage_final = damage_final.drop(columns = ['Total Affected'])

# first we drop the column
```

```python
damage_final['Total Affected'] = damage_final['Total Deaths'] + damage_final['No Injured'] + damage_final['No Homeless']

# Then we create it again which will be sum of three columns
```

```python
cols = list(damage_final.columns)
# Make our headings into a list

damage_final = damage_final[cols[0:17] + [cols[-1]] + [cols[17]]]
# Through this we can rearrange our columns to that total affected is not at the end
```

```python
damage_final = damage_final.sort_values('Year')
# Sort the dataframe by year
```

```python
#damage_final = damage_final.drop(columns = ['Disaster Subsubtype'])

damage_final = damage_final.drop(columns = ['Local Time'])
#Drop Local time and Disaster Subsubtype as they are not useful to us
```

```python
# We will check if any rows have NAN values

check = damage_final["Latitude"].isna()
damage_final[check]

# We can see that one row has NAN value in Longitude and Latitude.
# We will simply remove this one row as we cannot take average or fill the cells
```

```python
damage_final=damage_final.dropna(subset=['Latitude','Longitude'])

# Now all NAN values have been removed.
```
